```python
#Raskite didžiausią ir mažiausią įvestus atlyginimus.
import csv

rows = []

with open('2018.csv', encoding='utf8') as file:
    reader = csv.reader(file)
    headers = next(reader)
    stulpelio_indeksas = headers.index('Atlyginimas')
    for row in reader:
        rows.append(row)

stulpelio_reiksmes = [float(row[stulpelio_indeksas]) for row in rows]


minimalus_atlyginimas = min(stulpelio_reiksmes)


maksimalus_atlyginimas = max(stulpelio_reiksmes)
print(f'Maksimalus atlyginimas: {maksimalus_atlyginimas}, minimalus atlyginimas: {minimalus_atlyginimas}')
```

```python
#Raskite atlyginimų vidurkį.

import csv

# Read the file into a list of lists
eilutes = []
with open('2018.csv', encoding='utf8') as file:
    reader = csv.reader(file)
    headers = next(reader)
    stulpelio_indeksas = headers.index('Atlyginimas')
    for eilute in reader:
        eilutes.append(eilute)


reiksmes_stulpelyje = [float(row[stulpelio_indeksas]) for row in eilutes]


algos_vidurkis = sum(reiksmes_stulpelyje) / len(reiksmes_stulpelyje)

print(f'Vidutinis programuotoju atlyginimas: {algos_vidurkis}')
```

```python
#Raskite atlyginimų vidurkius pagal lygį (1 - junior, 2 - mid, 3 - senior).

import pandas as pd


df = pd.read_csv('2018.csv', encoding='utf8')

sugrupuota = df.groupby('lygis')

vidutinis_atlyginimas = sugrupuota['Atlyginimas'].mean()

print(vidutinis_atlyginimas)
```
